```python
def get_proxy():
    try:
        content = request.urlopen(
            "https://www.proxynova.com/proxy-server-list/country-tw/").read()
    except:
        raise AssertionError('Cannot access proxys now, plz try later!')
    soup = BeautifulSoup(content,'lxml')
    proxys = soup.find_all('tr',{'data-proxy-id':re.compile('\d{1,9}')})
    proxys_list = dict(zip(['{}:{}'.format(i.find_all('td')[0].find('abbr').attrs['title'].split('.')[0].replace('-','.'),i.find_all('td')[1].text.strip()) for i in proxys],[int(re.match('\d{1,9}',i.find_all('td')[3].text.strip()).group()) for i in proxys]))
    if proxys_list:
        return random.sample([i for i in proxys_list if proxys_list[i]==max(proxys_list.values())],1)[0]
    else:
        raise AssertionError('No TW proxy is found')
```

```python
proxies = {}
extractor_proxy = get_proxy()
proxies['http'] = 'http://{}'.format(extractor_proxy)
print("Using TW proxy {}".format(extractor_proxy))
```
